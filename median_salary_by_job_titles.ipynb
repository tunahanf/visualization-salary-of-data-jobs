```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.dropna(subset="salary_year_avg", inplace=True) # dropping NaN values.
```

```python
jobs = df.groupby("job_title_short")[["salary_year_avg"]].median().sort_values(by="salary_year_avg", ascending=True) # grouping values
```

```python
jobs.plot(kind="barh", legend=False)
plt.ylabel("Job Titles") # naming y axis
plt.xlabel("Salary ($USD)") # naming x axis
plt.title("Median Salary by Job Title") # naming chart
plt.show()
```
